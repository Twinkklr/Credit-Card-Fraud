3.0 Conclusions and Recommendations

This analysis demonstrates that a two-stage modeling approach is well suited to fraud detection in highly imbalanced transaction data. By separating the problem into an initial high-recall screening stage followed by a precision-focused refinement stage, the system is able to balance fraud detection effectiveness with the need to minimize unnecessary customer impact.

At the selected operating points, the system automatically intercepts a large majority of fraudulent transactions while incorrectly blocking only a very small fraction of legitimate activity. Transactions that fall near the decision boundary are flagged for customer contact or step-up verification rather than being blocked outright, allowing additional fraud prevention without materially increasing friction for most customers. This design reflects how fraud detection systems are typically deployed in practice, where automated decisions are complemented by downstream intervention rather than relying on a single hard threshold.

It is important to note that the dataset used in this assessment contains only transaction-level features and does not include account identifiers or transaction sequences. As a result, the models operate on individual transactions in isolation and cannot directly capture behavioral patterns such as transaction escalation or repeated probing activity. Despite this limitation, the results are consistent with known fraud behaviors and highlight the value of early, conservative screening combined with more targeted follow-up analysis.

From a financial perspective, the model captures a substantial portion of fraudulent activity while maintaining tight control over false positives. Although not all fraudulent transactions are automatically blocked, the system is designed to surface higher-risk transactions early, enabling additional actions that may prevent downstream losses even when individual transactions are not directly intercepted.

Based on these findings, the recommended deployment strategy would be to retain the two-stage structure, with conservative thresholds for automatic blocking and clearly defined criteria for customer outreach or step-up authentication. If additional data were available, future improvements could include incorporating account-level transaction history, optimizing explicitly for financial loss rather than transaction count, or introducing amount-aware decision rules for high-value transactions.

Overall, the results indicate that the proposed approach provides a practical and defensible balance between fraud prevention, operational efficiency, and customer experience.
